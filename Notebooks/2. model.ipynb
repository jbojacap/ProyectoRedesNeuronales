```python
import tensorflow as tf
from sklearn.preprocessing import OneHotEncoder, StandardScaler
from sklearn.compose import ColumnTransformer

# Seleccionar las características y el label
features = [
    'dia_semana', 'hora', 'numero_estacion', 'latitud_parada', 'longitud_parada',
    'latitud_parada0', 'longitud_parada0', 'latitud_parada25', 'longitud_parada25',
    'latitud_parada50', 'longitud_parada50', 'latitud_parada75', 'longitud_parada75',
    'latitud_parada100', 'longitud_parada100'
]
label = 'estimacionOcupacionAbordo'

# Separar características y etiquetas en los conjuntos de entrenamiento y prueba
X_train = train[features]
y_train = train[label]
X_test = test[features]
y_test = test[label]

# Configurar el preprocesador
preprocessor = ColumnTransformer(
    transformers=[
        ('onehot', OneHotEncoder(), ['dia_semana']),  # One-hot encoding para dia_semana
        ('scaler', StandardScaler(), [col for col in features if col != 'dia_semana'])  # Escalar numéricas
    ],
    remainder='passthrough'
)

# Preprocesar los datos
X_train_processed = preprocessor.fit_transform(X_train)
X_test_processed = preprocessor.transform(X_test)

# Crear el modelo de la red neuronal
model = tf.keras.Sequential([
    tf.keras.layers.Input(shape=(X_train_processed.shape[1],)),
    tf.keras.layers.Dense(64, activation='relu'),
    tf.keras.layers.Dense(32, activation='relu'),
    tf.keras.layers.Dense(1)  # Salida para estimación de ocupación
])

# Compilar el modelo
model.compile(optimizer='adam', loss='mse', metrics=['mae'])

# Entrenar el modelo
history = model.fit(
    X_train_processed, y_train,
    validation_data=(X_test_processed, y_test),
    epochs=10, batch_size=32, verbose=1
)

# Mostrar el resumen del modelo
model.summary()
```

```python
import matplotlib.pyplot as plt
import numpy as np
from scipy.stats import gaussian_kde

# Obtener las predicciones del modelo
y_pred_train = model.predict(X_train_processed).flatten()  # Predicciones para train
y_pred_test = model.predict(X_test_processed).flatten()  # Predicciones para test

# Calcular densidad para el conjunto de entrenamiento
xy_train = np.vstack([y_train, y_pred_train])
z_train = gaussian_kde(xy_train)(xy_train)

# Calcular densidad para el conjunto de prueba
xy_test = np.vstack([y_test, y_pred_test])
z_test = gaussian_kde(xy_test)(xy_test)

# Crear gráfico para el conjunto de entrenamiento
plt.figure(figsize=(10, 6))
plt.tricontourf(y_train, y_pred_train, z_train, levels=20, cmap='Blues')  # Curvas de nivel para train
plt.colorbar(label='Densidad')
max_val_train = max(y_train.max(), y_pred_train.max())
plt.plot([0, max_val_train], [0, max_val_train], color='red', linestyle='--', label='Línea Ideal')
plt.title('Densidad de Valores Reales vs. Predicciones (Train)', fontsize=14)
plt.xlabel('Valores Reales (Train)', fontsize=12)
plt.ylabel('Predicciones (Train)', fontsize=12)
plt.xlim(0, max_val_train)
plt.ylim(0, max_val_train)
plt.legend(fontsize=12)
plt.grid(alpha=0.3)
plt.tight_layout()
plt.show()

# Crear gráfico para el conjunto de prueba
plt.figure(figsize=(10, 6))
plt.tricontourf(y_test, y_pred_test, z_test, levels=20, cmap='Greens')  # Curvas de nivel para test
plt.colorbar(label='Densidad')
max_val_test = max(y_test.max(), y_pred_test.max())
plt.plot([0, max_val_test], [0, max_val_test], color='red', linestyle='--', label='Línea Ideal')
plt.title('Densidad de Valores Reales vs. Predicciones (Test)', fontsize=14)
plt.xlabel('Valores Reales (Test)', fontsize=12)
plt.ylabel('Predicciones (Test)', fontsize=12)
plt.xlim(0, max_val_test)
plt.ylim(0, max_val_test)
plt.legend(fontsize=12)
plt.grid(alpha=0.3)
plt.tight_layout()
plt.show()
```

```python
from sklearn.metrics import mean_absolute_error, mean_squared_error

# Calcular las métricas para el conjunto de prueba
mae_test = mean_absolute_error(y_test, y_pred_test)
mse_test = mean_squared_error(y_test, y_pred_test)
rmse_test = np.sqrt(mse_test)

# Crear un diccionario con las métricas
metrics_test = {
    'Mean Absolute Error (MAE)': mae_test,
    'Mean Squared Error (MSE)': mse_test,
    'Root Mean Squared Error (RMSE)': rmse_test
}

metrics_test
```
