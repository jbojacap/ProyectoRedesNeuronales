```python
# Realizar el inner join entre rutas_paradas y estaciones
# Realizar un join para agregar las coordenadas de 'NombreParada'
rutas_paradas_coordenadas = rutas_paradas.merge(
    estaciones,
    left_on='NombreParada',  # Unión con 'NombreParada'
    right_on='nombre_estacion',
    how='inner'  # Usamos left join por si no todas las paradas están en estaciones
).rename(columns={'latitud_estacion': 'latitud_parada', 'longitud_estacion': 'longitud_parada'})

# Realizar un join para agregar las coordenadas de 'estación_final'
rutas_paradas_coordenadas = rutas_paradas_coordenadas.merge(
    estaciones,
    left_on='estación_final',  # Unión con 'estación_final'
    right_on='nombre_estacion',
    how='inner'  # Usamos left join por si no todas las estaciones están en el dataset
).rename(columns={'latitud_estacion': 'latitud_estacion_final', 'longitud_estacion': 'longitud_estacion_final'})

# Eliminar columnas duplicadas innecesarias
rutas_paradas_coordenadas.drop(columns=['nombre_estacion_x', 'nombre_estacion_y'], inplace=True)

# Mostrar el resultado
rutas_paradas_coordenadas
```

```python
import numpy as np
import pandas as pd

# Función para calcular la distancia euclidiana
def calcular_distancia(p1, p2):
    return np.sqrt((p1[0] - p2[0])**2 + (p1[1] - p2[1])**2)

# Crear una columna inicial para el número de estación
rutas_paradas_coordenadas['numero_estacion'] = np.nan

# Iterar por cada ruta única
resultado = []
for ruta, grupo in rutas_paradas_coordenadas.groupby('nombre_ruta'):
    grupo = grupo.copy()
    
    # Encontrar el registro inicial (estación_final coincide con NombreParada)
    inicio = grupo[grupo['estación_final'] == grupo['NombreParada']].iloc[0]
    estacion_actual = (inicio['latitud_estacion_final'], inicio['longitud_estacion_final'])
    grupo.loc[inicio.name, 'numero_estacion'] = 1  # Asignar posición inicial
    
    # Resto de las estaciones
    posiciones_restantes = grupo[~(grupo.index == inicio.name)]
    numero = 2
    
    while not posiciones_restantes.empty:
        # Calcular la distancia desde la estación actual a todas las paradas restantes
        posiciones_restantes['distancia'] = posiciones_restantes.apply(
            lambda x: calcular_distancia(
                estacion_actual,
                (x['latitud_parada'], x['longitud_parada'])
            ),
            axis=1
        )
        
        # Seleccionar el registro con la menor distancia
        siguiente_indice = posiciones_restantes['distancia'].idxmin()
        grupo.loc[siguiente_indice, 'numero_estacion'] = numero
        
        # Actualizar la estación actual
        estacion_actual = (
            posiciones_restantes.loc[siguiente_indice, 'latitud_parada'],
            posiciones_restantes.loc[siguiente_indice, 'longitud_parada']
        )
        
        # Eliminar el registro seleccionado de los pendientes
        posiciones_restantes = posiciones_restantes.drop(index=siguiente_indice)
        numero += 1
    
    # Guardar el resultado para esta ruta
    resultado.append(grupo)

# Combinar los resultados de todas las rutas
rutas_paradas_secuenciadas = pd.concat(resultado)

# Mostrar el resultado final
print(rutas_paradas_secuenciadas)
```

```python
# Invertir el orden de las estaciones dentro de cada ruta
rutas_paradas_secuenciadas['numero_estacion'] = (
    rutas_paradas_secuenciadas.groupby('nombre_ruta')['numero_estacion']
    .transform(lambda x: x.max() - x + 1)  # Invertir el orden
)

# Mostrar el resultado final con el orden invertido
rutas_paradas_secuenciadas
```

```python
import folium

# Función para crear el mapa con la secuencia de rutas
def visualizar_rutas(data, ruta_seleccionada):
    # Filtrar por la ruta seleccionada
    ruta = data[data['nombre_ruta'] == ruta_seleccionada].sort_values(by='numero_estacion')
    
    # Crear el mapa centrado en la primera estción de la secuencia
    inicio = (ruta.iloc[0]['latitud_parada'], ruta.iloc[0]['longitud_parada'])
    mapa = folium.Map(location=inicio, zoom_start=13)
    
    # Añadir puntos al mapa y conectar con líneas
    coordenadas = []
    for _, row in ruta.iterrows():
        # Coordenadas de la parada
        parada = (row['latitud_parada'], row['longitud_parada'])
        coordenadas.append(parada)
        
        # Añadir un marcador para cada parada
        folium.Marker(
            location=parada,
            popup=f"Parada: {row['NombreParada']}<br>Estación Final: {row['estación_final']}<br>Número Estación: {int(row['numero_estacion'])}",
            tooltip=f"Número Estación: {int(row['numero_estacion'])}",
            icon=folium.Icon(color="blue", icon="info-sign")
        ).add_to(mapa)
    
    # Dibujar la línea que conecta las paradas en la secuencia
    folium.PolyLine(
        coordenadas,
        color="red",
        weight=2.5,
        opacity=0.8
    ).add_to(mapa)
    
    return mapa

# Seleccionar la ruta que deseas visualizar
nombre_ruta_ejemplo = "C30 Portal Suba"  # Cambia por el nombre de la ruta que quieras visualizar
mapa_ruta = visualizar_rutas(rutas_paradas_secuenciadas, nombre_ruta_ejemplo)

# Mostrar el mapa
mapa_ruta
```

```python
def calcular_percentiles(grupo):
    # Ordenar por numero_estacion
    grupo = grupo.sort_values('numero_estacion').reset_index(drop=True)
    
    # Calcular posiciones de los percentiles
    total_estaciones = len(grupo)
    percentiles = {
        'p0': 0,
        'p25': int(total_estaciones * 0.25),
        'p50': int(total_estaciones * 0.50),
        'p75': int(total_estaciones * 0.75),
        'p100': total_estaciones - 1
    }
    
    # Extraer coordenadas para cada percentil
    resultado = {
        'nombre_ruta': grupo['nombre_ruta'].iloc[0],
        'latitud_parada0': grupo.loc[percentiles['p0'], 'latitud_parada'],
        'longitud_parada0': grupo.loc[percentiles['p0'], 'longitud_parada'],
        'latitud_parada25': grupo.loc[percentiles['p25'], 'latitud_parada'],
        'longitud_parada25': grupo.loc[percentiles['p25'], 'longitud_parada'],
        'latitud_parada50': grupo.loc[percentiles['p50'], 'latitud_parada'],
        'longitud_parada50': grupo.loc[percentiles['p50'], 'longitud_parada'],
        'latitud_parada75': grupo.loc[percentiles['p75'], 'latitud_parada'],
        'longitud_parada75': grupo.loc[percentiles['p75'], 'longitud_parada'],
        'latitud_parada100': grupo.loc[percentiles['p100'], 'latitud_parada'],
        'longitud_parada100': grupo.loc[percentiles['p100'], 'longitud_parada']
    }
    
    return pd.DataFrame([resultado])

# Agrupar por nombre_ruta y calcular percentiles
resumen_rutas = (
    rutas_paradas_secuenciadas.groupby('nombre_ruta')
    .apply(calcular_percentiles)
    .reset_index(drop=True)
)

# Mostrar el DataFrame final
resumen_rutas
```

```python
import folium

# Función para crear el mapa con los puntos clave de la ruta
def visualizar_puntos_clave(data, ruta_seleccionada):
    # Filtrar por la ruta seleccionada
    ruta = data[data['nombre_ruta'] == ruta_seleccionada].iloc[0]  # Selecciona la fila correspondiente
    
    # Crear el mapa centrado en el primer punto (p0)
    inicio = (ruta['latitud_parada0'], ruta['longitud_parada0'])
    mapa = folium.Map(location=inicio, zoom_start=13)
    
    # Puntos clave
    puntos_clave = {
        'p0': (ruta['latitud_parada0'], ruta['longitud_parada0']),
        'p25': (ruta['latitud_parada25'], ruta['longitud_parada25']),
        'p50': (ruta['latitud_parada50'], ruta['longitud_parada50']),
        'p75': (ruta['latitud_parada75'], ruta['longitud_parada75']),
        'p100': (ruta['latitud_parada100'], ruta['longitud_parada100'])
    }
    
    # Añadir puntos al mapa y conectar con líneas
    coordenadas = []
    for clave, coordenada in puntos_clave.items():
        coordenadas.append(coordenada)
        
        # Añadir un marcador para cada punto
        folium.Marker(
            location=coordenada,
            popup=f"{clave}: {coordenada}",
            tooltip=f"Punto {clave}",
            icon=folium.Icon(color="blue", icon="info-sign")
        ).add_to(mapa)
    
    # Dibujar la línea que conecta los puntos clave
    folium.PolyLine(
        coordenadas,
        color="red",
        weight=2.5,
        opacity=0.8
    ).add_to(mapa)
    
    return mapa

# Seleccionar la ruta que deseas visualizar
nombre_ruta_ejemplo = "C30 Portal Suba"  # Cambia por el nombre de la ruta que quieras visualizar
mapa_ruta_clave = visualizar_puntos_clave(resumen_rutas, nombre_ruta_ejemplo)

# Mostrar el mapa
mapa_ruta_clave
```
